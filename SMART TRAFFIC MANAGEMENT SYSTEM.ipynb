```python
class TrafficDevice:
    def activate(self):
        print("Traffic Device Activated")

class TrafficLight(TrafficDevice):
    def activate(self):
        print("Traffic Light:Green light ON")

class SpeedCamera(TrafficDevice):
    def activate(self):
        print("Speed Camera Activated")

class PedestrianSignal(TrafficDevice):
    def activate(self):
        print("Pedestrian Signal Activated")

class EmergencySiren(TrafficDevice):
    def activate(self):
        print("Emergency Siren Activated")

devices = [
    TrafficLight(),
    SpeedCamera(),
    PedestrianSignal(),
    EmergencySiren(),
]
for device in devices:
    device.activate()
```
